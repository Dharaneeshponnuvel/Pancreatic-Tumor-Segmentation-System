```python
import os
import glob
import pandas as pd
import random
```

```python
# ---- CONFIG ----
images_dir = "K:\dharaneesh\dharaneesh\PANTHER_Task1\ImagesTr"
labels_dir = "K:\dharaneesh\dharaneesh\PANTHER_Task1\LabelsTr"
unlabeled_dir = "K:\dharaneesh\dharaneesh\PANTHER_Task1\ImagesTr_unlabeled"
# ---- CONFIG (Task2 paths) ----
task2_images_dir = "K:\dharaneesh\dharaneesh\PANTHER_Task2\ImagesTr"
task2_labels_dir = "K:\dharaneesh\dharaneesh\PANTHER_Task2\LabelsTr"


save_dir = "./csv_lists"
os.makedirs(save_dir, exist_ok=True)
```

```python
# ---- Helper: Create CSVs for labeled data ----
def make_train_val_csv(images_dir, labels_dir, split_ratio=0.8, shuffle=True, prefix="task1"):
    images = sorted(glob.glob(os.path.join(images_dir, "*.mha")))
    labels = sorted(glob.glob(os.path.join(labels_dir, "*.mha")))
    assert len(images) == len(labels), "Mismatch: images and labels count differ"

    data = list(zip(images, labels))
    if shuffle:
        random.shuffle(data)

    split = int(len(data) * split_ratio)
    train, val = data[:split], data[split:]

    pd.DataFrame(train).to_csv(os.path.join(save_dir, f"{prefix}_train.csv"), index=False, header=False)
    pd.DataFrame(val).to_csv(os.path.join(save_dir, f"{prefix}_val.csv"), index=False, header=False)

    print(f"Created {prefix}_train.csv ({len(train)} pairs) and {prefix}_val.csv ({len(val)} pairs)")
```

```python
# ---- Helper: Create CSV for unlabeled images ----
def make_unlabeled_csv(unlabeled_dir, prefix="task1_unlabeled"):
    images = sorted(glob.glob(os.path.join(unlabeled_dir, "*.mha")))
    pd.DataFrame(images).to_csv(os.path.join(save_dir, f"{prefix}.csv"), index=False, header=False)
    print(f"Created {prefix}.csv ({len(images)} images)")
```

```python
# ---- Run ----
make_train_val_csv(images_dir, labels_dir, split_ratio=0.8, prefix="task1")
make_unlabeled_csv(unlabeled_dir, prefix="task1_unlabeled")
make_train_val_csv(task2_images_dir, task2_labels_dir, split_ratio=0.8, prefix="task2")
```
